```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import sklearn
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import learning_curve

from feature_engine.datetime import DatetimeFeatures

from xgboost import XGBRegressor

import joblib
```

```python
pd.set_option("display.max_columns", None)
```

```python
sklearn.set_config(transform_output="pandas")
```

```python
train_df = pd.read_csv("data/train.csv")
val_df = pd.read_csv("data/val.csv")
test_df = pd.read_csv("data/test.csv")
```

```python
train_df
```

```python
def split_data(data):
    X = data.drop(columns = "price")
    y = data.price.copy()
    return (X,y)
```

```python
X_train, y_train = split_data(train_df)
```

```python
X_train
```

```python
X_val, y_val = split_data(val_df)

print(X_val.shape, y_val.shape)
```

```python
X_test, y_test = split_data(test_df)

print(X_test.shape, y_test.shape)
```

```python
X_train.info()
```

```python

dt_cols  = ["date_of_journey", "dep_time", "arrival_time"]
num_cols = ["total_stops", "duration"]
cat_cols = [col for col in X_train.columns if (col not in dt_cols) and (col not in num_cols)]
```

```python
cat_cols
```

```python
num_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(sparse_output=False))
])

doj_transformer = Pipeline(steps=[
    ("impter", SimpleImputer(strategy="most_frequent")),
    ("extracter", DatetimeFeatures(features_to_extract=["month", "week", "day_of_week", "day_of_month"], format ="mixed")),
    ("scaler", StandardScaler())
])

time_transformer = Pipeline(steps=[
    ("impter", SimpleImputer(strategy="most_frequent")),
    ("extracter", DatetimeFeatures(features_to_extract=["hour", "minute"], format ="mixed")),
    ("scaler", StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", num_transformer, num_cols),
    ("cat", cat_transformer, cat_cols),
    ("doj", doj_transformer, ["date_of_journey"]),
    ("time", time_transformer, ["dep_time", "arrival_time"])
])
```

```python
preprocessor.fit_transform(X_train)
```

```python
algorithms = {
    "Linear Regression" : LinearRegression(),
    "Support Vector Machine": SVR(),
    "Random Forest":RandomForestRegressor(n_estimators=10),
    "XG Boost":XGBRegressor(n_estimators=10)
}
```

```python
data = pd.concat([train_df, val_df], axis = 0)

X_data, y_data = split_data(data)
print(X_data.shape, y_data.shape)
```

```python
def plot_curves(sizes, mean_scores, std_scores, label, ax):
    ax.plot(sizes, mean_scores, marker="o", label=label)
    ax.fill_between(sizes, mean_scores - std_scores, mean_scores + std_scores, alpha=0.5)
```

```python
def plot_learning_curve(name, algorithm, preprocessor, X_data, y_data, figsize=(12, 4)):
    model = Pipeline(steps=[
        ("pre", preprocessor),
        ("alg", algorithm)
    ])

    train_sizes, train_scores, test_scores = learning_curve(
        estimator=model,
        X=X_data,
        y=y_data,
        cv=3,
        scoring="r2",
        n_jobs=-1
    )

    mean_train_scores = np.mean(train_scores, axis=1)
    std_train_scores = np.std(train_scores, axis=1)
    train_label = f"{mean_train_scores[-1]:.2f} ± {std_train_scores[-1]:.2f}"

    mean_test_scores = np.mean(test_scores, axis=1)
    std_test_scores = np.std(test_scores, axis=1)
    test_label = f"{mean_test_scores[-1]:.2f} ± {std_test_scores[-1]:.2f}"

    fig, ax = plt.subplots(figsize=figsize)

    plot_curves(train_sizes, mean_train_scores, std_train_scores, f"Train ({train_label})", ax)
    plot_curves(train_sizes, mean_test_scores, std_test_scores, f"Test ({test_label})", ax)

    ax.set(xlabel="Training Set Size", ylabel="R² Score", title=name)
    ax.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
```

```python
for name, alg in algorithms.items():
    plot_learning_curve(name, alg, preprocessor, X_data, y_data)
```

```python
model = Pipeline(steps=[
    ("pre", preprocessor),
    ("ref", RandomForestRegressor(n_estimators=25))
])
```

```python
model.fit(X_data, y_data)
```

```python
def evaluate_model(X, y):
    y_pred = model.predict(X)
    return r2_score(y, y_pred)
```

```python
print(f"R2 score on Training data is = {evaluate_model(X_data, y_data)}")
```

```python
print(f"R2 score on Test data is = {evaluate_model(X_test, y_test)}")
```
